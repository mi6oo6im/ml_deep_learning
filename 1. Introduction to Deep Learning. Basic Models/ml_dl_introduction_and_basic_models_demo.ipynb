$$ (2a + 3b)^2 $$

```python
# plain vanilla Python
a, b = 15, 20
(2 * a + 3 * b) ** 2
# 8100
```

```python
# numpy arrays
a, b = np.array([15]), np.array([20])
(2 * a + 3 * b) ** 2
# array([8100.])
```

```python
# TensorFlow
a, b = tf.constant(15), tf.constant(20)
(2 * a + 3 * b) ** 2
# tf.Tensor: shape=(), dtype=int32, numpy=8100
```

```python
# PyTorch
a, b = torch.tensor(15), torch.tensor(20)
(2 * a + 3 * b) ** 2
# tensor(8100)
```

```python
weights = torch.randn(10, 5, requires_grad=True)
intercept = torch.randn(5, requires_grad=True)

activation = torch.matmul(x, weights) + intercept
```
